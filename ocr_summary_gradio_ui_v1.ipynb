```python
# Here we import the required libraries for OCR and transformers models

from transformers import BartTokenizer, BartForConditionalGeneration
import textwrap
from PIL import Image
import pytesseract
from transformers import pipeline
import os
import math
import gradio as gr

pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')

```

```python
# Return the most recent file (the downloaded image)

def return_recent_file():

    # List all files in the directory
    files = os.listdir('./')

    # Filter out only files (excluding subdirectories)
    files = [file for file in files if os.path.isfile(os.path.join('./', file))]

    # Get the most recent file based on modification time
    if files:
        most_recent_file = max(files, key=lambda x: os.path.getmtime(os.path.join('./', x)))
    else:
        print("No files found in the directory.")

    return most_recent_file
```

```python
# Function that converts the image to jpg, scans with OCR and returns text

def begin_ocr(scan_img):

    scan_img.save("./converted_image.jpg")

    recent_file  = return_recent_file()

    image = Image.open('./' + recent_file)

    # Perform OCR using pytesseract
    text = pytesseract.image_to_string(scan_img)

    return text

# Function that summarises the text with an LLM, summary strength can be controlled.

def summarizer_func(text, sum_strength):

  input_tokens = tokenizer.batch_encode_plus([text], return_tensors='pt', max_length = 1024, truncation = True)['input_ids']

  encoded_ids = model.generate(input_tokens,
                              num_beams=4,
                              length_penalty=2.0,
                              max_length=sum_strength[1],
                              min_length = sum_strength[0],
                              no_repeat_ngram_size=3)

  summary = tokenizer.decode(encoded_ids.squeeze(), skip_special_tokens = True)


  print(textwrap.fill(summary, 100))

  return textwrap.fill(summary, 100)

# Main function process workflow

def convert_doc(scan_img, bool_summary, sum_strength_sel):

  text = begin_ocr(scan_img)

  temp_sum_chunks_list = []

  if bool_summary:

      total_words = len(text.split())

      # The strength is controlled with a slider, output will be a ratio of total words.

      sum_strength = [[round(total_words*0.6), round(total_words*0.8)],
                      [round(total_words*0.4), round(total_words*0.6)],
                      [round(total_words*0.25), round(total_words*0.4)],
                      [round(total_words*0.15), round(total_words*0.25)]][round(sum_strength_sel)-1]

  return text

```

```python
# Build and launch Gradio Web-UI

demo = gr.Interface(
    convert_doc,

    [gr.Image(type="pil"),
     gr.Checkbox(label="Enable Summary", info="An LLM will summarise the image text", value=False),
     gr.Slider(1, 4, label="Strength of Summary", info="Higher the value, the more concise the summary", value=2),
    ],
    outputs = "text",
)

demo.launch(debug=True)
```
